```python
import pandas as pd
import numpy as np
```

```python
def resumen(datos):
    return print('Forma:\n', datos.shape,'\n\n',
                 'Columnas:\n', datos.columns,'\n\n',
                 'Tipos:\n', datos.dtypes,'\n\n',
                 'Datos nulos:\n', datos.isnull().sum(),'\n\n',
                 'Cabecera:\n', datos.head(10),'\n\n',
                 'Últimos datos:\n', datos.tail(10),'\n\n',
                 'Estadísticos:\n', datos.describe([0.05,0.25,0.5,0.75,0.95])
                )
def truncar(numero, decimales = 0):
    return np.trunc(numero*10**decimales)/(10**decimales)
```

```python
datos_juntos = pd.read_csv(
    'C:\\Users\\nahue\\Desktop\\Tesis_2\\Datos\\Datos_buenos_2.csv',
    delimiter = '\t',
    decimal = '.',
    na_values = -9999,
    encoding = 'ascii'
    )
```

```python
datos_para_promediar = datos_juntos[['anio', 'Dia', 'Hora', '  NEE[mmol/m2s]',
                                     'LvE[W/m2]      ', 'HTs[W/m2]      ', 'RG',
                                     'TA1', 'TSS1', 'HR1', 'VPD 1',  'TSS2',
                                     'ustar[m/s]     '
                                ]]
```

```python
resumen(datos_para_promediar)
```

```python
datos_para_promediar['  NEE[mmol/m2s]'] = datos_para_promediar['  NEE[mmol/m2s]'] * 1000
```

```python
datos_para_promediar['anio'] = 2018
```

```python
datos_para_promediar['TSS1'] = ((datos_para_promediar['TSS1']
                                + datos_para_promediar['TSS2']
                                ) / 2
                               )
```

```python
unidades = ['-', '-', '-', 
            'umolm-2s-1', 'Wm-2', 
            'Wm-2', 'Wm-2', 'degC',
            'degC', '%', 'hPa','ms-1'
            ]
columnas = {'anio':'Year',
            'Dia':'DoY',
            'Hora':'Hour',
            '  NEE[mmol/m2s]':'NEE',
            'LvE[W/m2]      ':'LE',
            'HTs[W/m2]      ':'H',
            'RG':'Rg',
            'TA1':'Tair',
            'TSS1':'Tsoil',
            'HR2':'rH',
            'VPD 1':'VPD',
            'ustar[m/s]     ':'Ustar'
           }
```

```python
datos_para_promediar = datos_para_promediar.rename(columns = columnas)
```

```python
datos_para_promediar = datos_para_promediar.drop('TSS2', axis = 1)
```

```python
datos_para_promediar.head()
```

```python
datos_promediados = datos_para_promediar[1:].rolling(2, min_periods = 1).mean()
datos_promediados.head(10)
```

```python
iterador = list(range(2, len(datos_promediados['Year'])))
datos_para_subir = datos_promediados[:]
for i in iterador:
    truncado = truncar(datos_promediados['Hour'][i])
    decimales = datos_promediados['Hour'][i] - truncado
    if decimales == 0.125:
        datos_para_subir['Hour'][i] = 0.5 + truncado
    elif decimales == 0.625:
        datos_para_subir['Hour'][i] = truncado + 1
        if datos_para_subir['Hour'][i] == 24:
            datos_para_subir['Hour'][i] = 0
            datos_para_subir['DoY'][i] = datos_para_subir['DoY'][i] + 1
    else: datos_para_subir = datos_para_subir.drop(i)
```

```python
datos_para_subir.loc[-1] = unidades
datos_para_subir.index = datos_para_subir.index + 1
datos_para_subir.sort_index(inplace = True)
```
