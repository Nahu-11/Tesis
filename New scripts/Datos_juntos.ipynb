```python
import pandas as pd
import numpy as np
```

```python
def resumen(datos):
    return print('Forma:\n', datos.shape,'\n\n',
                 'Columnas:\n', datos.columns,'\n\n',
                 'Tipos:\n', datos.dtypes,'\n\n',
                 'Datos nulos:\n', datos.isnull().sum(),'\n\n',
                 'Cabecera:\n', datos.head(10),'\n\n',
                 'Últimos datos:\n', datos.tail(10),'\n\n',
                 'Estadísticos:\n', datos.describe([0.05,0.25,0.5,0.75,0.95])
                )
```

```python
def truncar(numero, decimales = 0):
    return np.trunc(numero*10**decimales)/(10**decimales)
```

```python
datos_tk3 = pd.read_csv(
        'C:\\Users\\nahue\\Desktop\\Tesis_2\\Datos\\TK3.csv',
        delimiter = ',',
        decimal = '.',
        )
```

```python
columnas = ['Momento','TSS1','TSS2','TS1','TS2','TS3','TS4','TS_QG','U','DIR',
'PP', 'TA1','HR1','TA2','HR2','TA3','HR3','TA4','HR4','QG1','QG2','WET','RG','PAR',
'TSH1','HS1','GH1','Real(KH)1','Imag(KH)1','Real_T(KH)1','Imag_T(KH)1','TSH2',
'HS2','GH2','Real(KH)2','Imag(KH)2','Real_T(KH)2','Imag_T(KH)2','TSH3','HS3',
'GH3','Real(KH)3','Imag(KH)3','Real_T(KH)3','Imag_T(KH)3','TSH4','HS4','GH4',
'Real(KH)4','Imag(KH)4','Real_T(KH)4','Imag_T(KH)4','TSH5','HS5','GH5','Real(KH)5',
'Imag(KH)5','Real_T(KH)5','Imag_T(KH)5']
cant_col = len(columnas)
datos_conv = pd.read_csv(
        'C:\\Users\\nahue\\Desktop\\Tesis_2\\Datos\\Convencional.csv',
        delimiter = ';',
        decimal = '.',
        skiprows = [1,2],
        usecols = columnas,
        dtype = 'float64',
        engine = 'python'
        )
```

```python
resumen(datos_tk3)
```

```python
resumen(datos_conv)
```

```python
datos_conv['Dia'] = datos_conv['Momento'].apply(np.floor).astype(int)
datos_conv['Hora'] = (datos_conv['Momento']-datos_conv['Dia']) * 24
datos_conv['Hora'] = datos_conv['Hora'].round(2)
```

```python
datos_conv['Dia'].nunique()
```

```python
datos_conv['Hora'].nunique()
```

```python
longitud_hora = list(range(len(datos_tk3['T_begin        '])))
for i in longitud_hora:
    horario = int(datos_tk3['T_begin        '][i][11:13]
                + datos_tk3['T_begin        '][i][14:16]
               )
    if (horario/5) != int(horario/5):
        horario = horario + 8
    horario = horario/100
    if round(horario, 1) == 23.6:
        horario = 0
    minutos = (horario - np.floor(horario)) * 100/60
    longitud_hora[i] = truncar(np.floor(horario) + minutos,
                               1)
    decimas = round(longitud_hora[i] - np.floor(longitud_hora[i]),
                    1)
    if decimas == 0.2 or decimas == 0.7:
        longitud_hora[i] = longitud_hora[i] + 0.05
    if decimas == 0.3 or decimas == 0.8:
        longitud_hora[i] = longitud_hora[i] - 0.05
    if decimas == 0.9 or decimas == 0.4:
        longitud_hora[i] = longitud_hora[i] + 0.1
datos_tk3['Hora'] = longitud_hora
```

```python
datos_tk3['Hora'].sample(5)
```

```python
inicio_meses_julianos = {'01':0, '02':31, '03':59, '04':90,
                         '05':120, '06':151, '07':181, '08':212,
                         '09':243, '10':273, '11':304, '12':334}
longitud_fecha = list(range(len(datos_tk3['T_begin        '])))
for i in longitud_fecha:
    dia_gregoriano = datos_tk3['T_begin        '][i][0:2]
    mes_gregoriano = datos_tk3['T_begin        '][i][3:5]
    longitud_fecha[i] = (int(dia_gregoriano)
                         + inicio_meses_julianos[mes_gregoriano]
                         )
datos_tk3['Dia'] = longitud_fecha
```

```python
datos_tk3['Dia'].nunique()
```

```python
datos_juntos = pd.merge(datos_conv,
                        datos_tk3,
                        on = ['Dia', 'Hora'],
                        how = 'outer'
                       )
```

```python
resumen(datos_juntos)
```

```python
datos_juntos['Hora'].sample(5)
```

```python
datos_juntos['anio'] = datos_juntos['Hora']
datos_juntos.assign(anio = 2018)
```

```python
iterador = list(range(len(datos_juntos['Flag(ustar)    '])))
for i in iterador:
    if datos_juntos['Flag(ustar)    '][i] > 6:
        datos_juntos['ustar[m/s]     '][i] = np.nan
    if datos_juntos['Flag(HTs)      '][i] > 6:
        datos_juntos['HTs[W/m2]      '][i] = np.nan
    if datos_juntos['Flag(LvE)      '][i] > 6:
        datos_juntos['LvE[W/m2]      '][i] = np.nan
    if datos_juntos['Flag(wCO2)     '][i] > 6:
        datos_juntos['  NEE[mmol/m2s]'][i] = np.nan
```
